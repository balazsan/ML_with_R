<font size=6>Introduction to Machine Learning</font><br><br>

* <font size=5>Machine learning is the study of algorithms that <ins>learn from examples and experience</ins> instead of relying on <ins>hard-coded rules</ins> and make predictions on new data.</font><br><br>
* <font size=5>A computer program is said to learn from <ins>experience E</ins> with respect to some class of <ins>tasks T</ins> and <ins>performance measure P</ins>, if its performance at tasks in T, as measured by P, improves with experience E. (Tom M. Mitchell)</font>

* <font size=5>In the exercise target values are forest attributes (first table) and the data used to predict forest attributes are remote sensing features calculated from LiDAR data (second table).<br>The number of features (columns) can be very high, sometimes more than 1000. The data used in the exercise has 198 features.</font><br><br>
* <font size=5>In supervised learning we try to model a connection between x and y (task <i>T</i> according to Mitchell)</font><br><br>
* <font size=5>Types of supervised learning:</font>
    * <font size=5>classification (y is categorical (e.g. forest type))</font><br><br>
    * <font size=5>regression (y is continuous, real value (e.g. volume, height))<br>the exercise of this course is a regression problem</font><br><br>
* <font size=5>The "goodness" of the created model needs to be evaluated, this is the performance measure <i>P</i> according to Mitchell, which can be accuracy for classification problems and e.g. Mean Squared Error (MSE) for regressions</font><br><br>
* <font size=5>Performance is typically measured on a fraction of the data not included in the modelling phase; here we will call it test set</font>
